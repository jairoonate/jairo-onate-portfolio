```python
import pandas as pd
data = pd.read_csv('/content/OUData_cleaned_NONstandardized.csv')
data
```

```python
import matplotlib.pyplot as plt

plt.hist(data['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Age', fontsize=12, fontweight='bold')
plt.ylabel('Frequency', fontsize=12, fontweight='bold')
plt.xticks(fontsize=9, fontweight='bold')
plt.show()
```

```python
gender_counts = data['Gender'].value_counts()
gender_counts.plot(kind='bar', color=['pink', 'lightblue'])  # Custom colors for each gender
plt.title('Gender Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Gender', fontsize=12, fontweight='bold')
plt.ylabel('Count',fontsize=12, fontweight='bold')
plt.xticks(rotation=0)
plt.xticks(fontsize=9, fontweight='bold')
plt.show()
```

```python
plt.scatter(data['BloodPressureUpper'], data['BloodPressureLower'], c=data['BloodPressureUpper'], cmap='coolwarm', alpha=0.7)
plt.title('Blood Pressure Upper vs Lower', fontsize=14, fontweight='bold')
plt.xlabel('Blood Pressure Upper', fontsize=11, fontweight='bold')
plt.ylabel('Blood Pressure Lower', fontsize=11, fontweight='bold')
plt.colorbar(label='Blood Pressure Upper')
plt.show()
```

```python
plt.figure(figsize=(10,6))
data.boxplot(column='OU_LOS_hrs', by='InitPatientClassAndFirstPostOUClass', patch_artist=True,
             boxprops=dict(facecolor='skyblue', color='black'),
             medianprops=dict(color='red', linewidth=2))
plt.title('Length of Stay vs Patient Class', fontsize=14, fontweight='bold')
plt.suptitle('')
plt.xlabel('Patient Class', fontsize=11, fontweight='bold')
plt.ylabel('Length of Stay (hours)', fontsize=12, fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(data['Respirations'], data['PulseOximetry'], c=data['PulseOximetry'], cmap='viridis', alpha=0.6)
plt.title('Pulse vs Pulse Oximetry (Oxygen Levels)', fontsize=14, fontweight='bold')
plt.xlabel('Respirations', fontsize=11, fontweight='bold')
plt.ylabel('Pulse Oximetry', fontsize=11, fontweight='bold')
plt.colorbar(label='Pulse Oximetry')  # Color bar to show scale
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Select relevant variables
variables = ['Age', 'OU_LOS_hrs', 'BloodPressureUpper', 'BloodPressureLower', 'Pulse', 'PulseOximetry', 'Respirations', 'Temperature']

# Correlation Matrix (for relationships between variables)
correlation_matrix = data[variables].corr()

# Visualizing the correlation matrix with a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Matrix')
plt.show()
```
